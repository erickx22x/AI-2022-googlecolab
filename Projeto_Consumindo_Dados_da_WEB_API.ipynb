##Importando bibliotecas

```python
import pandas as pd
import requests
```

##Leitura do URL

Sons e localização de pássaros pelo Brasil

https://www.xeno-canto.org/api/2/recordings?query=cnt:brazil


```python
url = 'https://www.xeno-canto.org/api/2/recordings?query=cnt:brazil'
```

```python
r = requests.get(url)
```

##Extraindo dados usando Json

```python
json = r.json()
```

```python
json
```

```python
json.keys()
```

##JSON para Dataframe

```python
df = pd.DataFrame(json['recordings'])
```
